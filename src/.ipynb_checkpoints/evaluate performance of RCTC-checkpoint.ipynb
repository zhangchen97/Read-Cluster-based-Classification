## seq2Fasta

```python
def seq2fasta(seq_path,fasta_path):
    infile=open(seq_path,'r')
    outfile=open(fasta_path,'w+')
    for seq in infile:
        seq_lst=seq.split("\t")
        readid=seq_lst[2]
        read=seq_lst[3]
        outfile.write(">"+readid+'\n'+read)
    infile.close()
    outfile.close()
    print("finised")
```

```python
#inpath="/home/ubuntu/kraken2/database2/taxonomy/nodes.dmp"
#tmp_inpath="/home/ubuntu/kraken2res/tmp_nodes.dmp"
##get taxonomy's df，第一列是taxid,第二列是rank
##input:filepath，output:dataframe
def get_taxdf(inpath):
    infile=open(inpath,'r')
    rank=set()
    nodedict=dict()
    for line in infile:
        line=line.strip()
        arr=line.split("|")
        tmp_arr=[]
        for item in arr:
            tmp_arr.append(item.strip())
        nodedict[int(tmp_arr[0])]=str(tmp_arr[2])
        rank.add(tmp_arr[2])
    infile.close()
    print(rank)
    taxser=pd.Series(nodedict)
    taxdf=pd.DataFrame(taxser)
    taxdf=taxdf.reset_index(drop=False)
    taxdf.columns=['taxid','rank']
    print(taxdf.head())
    print(taxdf.shape)
    return taxdf

# inpath="/home/ubuntu/kraken2/compeleteres2.txt"
# labelpath="/home/ubuntu/mycami2/label50sample"
# get_kradf_cludf(inpath,labelpath)
##输入1:kraken2 taxonomy result file path，
##   2: reads的数量（注意，每个cluster的reads是连续的，而且数量一样多）,现在是50
##output: dataframe,第一列是readname,第二列是kra_id（不能保证rank）
def get_clark_df(inpath):
    df = pd.read_csv(inpath, sep=",",usecols=[0,2])
    df.columns=['readname','clark_id']
    df.columns=['readname','clark_id']
    df['clark_id'].fillna(0,inplace=True)
    print(df.head())
    print(df.shape)
    return df

#输入：1：raw read data(readname,seq)，
#     2：cami2的label data file（readname,label），
#     3：cami2 label2taxid mapping file(label,taxid)
def get_rawdf(inpath,labelpath,metapath,clusterpath):
    nameset=set()
    count=0
    infile=open(inpath,'r')
    for line in infile:
        if(count%2==0):
            readname=line.strip().replace(">","")
            nameset.add(readname)
        count=count+1
    infile.close()
    
    ##读取答案文件
    labelfile=open(labelpath,'r')
    dict1=dict()
    for line in labelfile:
        arr=line.strip().split("\t")
        if arr[0] in nameset:
            dict1[arr[0]]=arr[1]
    labelfile.close()
    ser=pd.Series(dict1)
    df=pd.DataFrame(ser,index=None)
    df=df.reset_index(drop=False)
    df.columns=['readname','labelid']

    ##读取label到taxid的映射
    meta = pd.read_csv(metapath, sep="\t",usecols=[0,2])
    meta.columns=['labelid','taxid']
    df=pd.merge(df,meta,how='inner',on='labelid')
    
    ##读取local clustering到read的映射
    clu_df = pd.read_csv(clusterpath, sep="\t",header=None)
    clu_df.columns=['readname','clusterid']
    df=pd.merge(df,clu_df,how='inner',on='readname')
    df=df.sort_values(by=['readname'])
    print("=======================================打印前五行===================================================")
    print(df.head())
    print(df.shape)
    return df

inpath="/home/ubuntu/kraken2/database2/taxonomy/nodes.dmp"
tmp_inpath="/home/ubuntu/kraken2res/tmp_nodes.dmp"
infile=open(inpath,'r')
rank=set()
nodedict=dict()
for line in infile:
    line=line.strip()
    arr=line.split("|")
    tmp_arr=[]
    for item in arr:
        tmp_arr.append(item.strip())
    #print(tmp_arr)
    nodedict[int(tmp_arr[0])]=[int(tmp_arr[1]),str(tmp_arr[2])]
    rank.add(tmp_arr[2])
infile.close()

## level选择species,genus, Obtain the taxid corresponding to the level
def get_level(taxid,level):
    if(taxid==0):  ##未分类
        return 0
    if taxid not in nodedict:
        return -1
    rank=nodedict[taxid]
    parent_taxid=rank[0]
    rank_name=rank[1]

    if(rank_name==level): ##本身就是level级别的
        return taxid
    
    above_level={"kingdom","phylum","class","family","order"}
    if(level=="species"): 
        above_level.add("genus")

    while( rank_name not in above_level):
        rank=nodedict[parent_taxid]
        rank_name=rank[1]
        #print(rank)
        if(rank_name==level):
            return parent_taxid
        if(parent_taxid==rank[0]): ##还等于上一次循环的taxid，因此来避免死循环
            return -1
        parent_taxid=rank[0]
    if(rank_name!=level):
        return -1

def get_mergealldf(df,taxdf,kra_df):
    #产生答案的df
    new_df=pd.merge(df,taxdf,how='left',on='taxid')
    new_df.rename(columns = {"rank": "true_rank","taxid": "true_taxid"},  inplace=True)
    print("all the rank of raw reads result:  ")
    print(set(list(new_df['true_rank'])))

    ##合并kraken2预测值
    new_df2=pd.merge(new_df,kra_df,how='left',on='readname')
    new_df2=pd.merge(new_df2,taxdf,how='left',left_on='clark_id',right_on='taxid')
    new_df2=new_df2.drop(['taxid'],axis=1)
    new_df2.rename(columns = {"rank": "clark_rank"},  inplace=True)
    print("all the rank of clark result: ")
    print(set(list(new_df2['clark_rank'])))
    print(new_df2.head())
    return new_df2

def Pre_Sen(lst_true,lst_pre): 
    numOfAmbiguousTrue=0;  ##代表在某个level上，真实值的taxID没找到，赋值为-1
    numOfAmbiguousPredictionTrue=0  ##在某个level上，真实值和预测值的taxID都没找到，赋值为-1
    numOfUnClassified=0   ##没有分类的数量
    numOfCorrectAssign=0;
    numOfAssign=0;
    numOfAll=len(lst_true);
    for i in range(len(lst_true)):
        if(lst_true[i]==-1):
            numOfAmbiguousTrue=numOfAmbiguousTrue+1
        if(lst_true[i]==-1 and lst_pre[i]==-1):##代表没有答案
            numOfAmbiguousPredictionTrue=numOfAmbiguousPredictionTrue+1
            continue
        if(lst_pre[i]==0):
            numOfUnClassified=numOfUnClassified+1
        if(lst_pre[i]!=0 and lst_pre[i]==lst_true[i]):
            numOfCorrectAssign=numOfCorrectAssign+1
            numOfAssign=numOfAssign+1;
        if(lst_pre[i]!=0 and lst_pre[i]!=lst_true[i]):
            numOfAssign=numOfAssign+1;
    print("numOfAmbiguousTrue: "+str(numOfAmbiguousTrue))
    print("numOfAmbiguousPredictionTrue: "+str(numOfAmbiguousPredictionTrue))
    print("numOfUnClassified: "+str(numOfUnClassified))
    print("numOfCorrectAssign: "+str(numOfCorrectAssign))
    print("numOfAssign: "+str(numOfAssign))
    print("numOfAll: "+str(numOfAll))
    print("Precision: "+str(numOfCorrectAssign/numOfAssign))
    print("Sensitivity: "+str(numOfCorrectAssign/numOfAll))
```

```python
def percent(values):
    fake=list(sort(values))
    SUM=float(reduce(lambda x,y:x+y,fake))
    j=0
    thd=np.power(10,np.linspace(1.69,7,54))
   # thd=[100,1000,10000,100000,1000000,10000000]
    pcd = [[0] * len(fake) for i in range(len(thd))]
    alist=[]
    for i in np.arange(len(fake)):
        if thd[j]<=fake[i]<thd[j+1]:
            pcd[j].append(fake[i])
        else:
            j=j+1
            pcd[j].append(fake[i])
    for j in np.arange(len(thd)):
        alist.append(sum(pcd[j])/SUM*100)
    return alist
def add_labels(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2, height, height, ha='center', va='bottom')
        rect.set_edgecolor('white')
```

```python
#输入：
#     1：cami2的label文件数据路径（readname,label），
#     2：cami2的label到taxid的映射文件 (label,taxid)
#     3: kraken2分类结果
#     4： taxonomy rank level: species
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (12, 8)
from matplotlib_venn import venn2
def get_AllReads_Tax_Venn2(labelpath,metapath,krapath,kra_row_all,level):
    ##读取答案文件
    labelfile=open(labelpath,'r')
    readLabelDict=dict()
    for line in labelfile:
        arr=line.strip().split("\t")
        readName=arr[0]
        labelId=arr[1]
        readLabelDict[readName]=labelId
    labelfile.close()
    print("readLabelDict")
    print(len(readLabelDict))
    
    metaFile=open(metapath,'r')
    labelNCBIID_Dict=dict()
    for line in metaFile:
        arr=line.strip().split("\t")
        label=arr[0]
        NCBI_ID=arr[2]
        labelNCBIID_Dict[label]=NCBI_ID
    metaFile.close()
    print("labelNCBIID_Dict")
    print(len(labelNCBIID_Dict))
    ##
    readNCBI_DICT=dict()
    for readName,labelId in readLabelDict.items():
        readNCBI_DICT[readName]=labelNCBIID_Dict[labelId]
    print("readNCBI_DICT")
    print(len(readNCBI_DICT))
    
    TrueSpeicesNCBISet=set()
    for readName,NCBIId in readNCBI_DICT.items():
        #get the species taxid
        species_taxid=get_level(int(NCBIId),level)
        TrueSpeicesNCBISet.add(int(species_taxid))
    print("TrueSpeicesNCBISet")
    print(len(TrueSpeicesNCBISet))
    
    krafile=open(krapath,"r")
    row_num=0
    kraSet=set()
    for line in krafile:
        row_num=row_num+1
        if(row_num<=18 or row_num>(kra_row_all-3)):
            continue
        arr=line.strip().split("\t")
        kraSet.add(arr[2])
    krafile.close()
    KraSpeciesSet=set()
    for taxid in kraSet:
        species_id=get_level(int(taxid),level)
        KraSpeciesSet.add(int(species_id))
    print("KraSpeciesSet")
    print(len(KraSpeciesSet))
    
    venn2(subsets=[TrueSpeicesNCBISet,KraSpeciesSet],set_labels=('True_'+level,'Kraken_'+level),set_colors=('r','g'))
    plt.show()
```

```python
def get_metaphlan_level(path,taxlevel):
    readfile=open(path,"r")
    taxmap=dict()
    count=0
    for line in readfile:
        count=count+1
        #if(count<=4 or count>532):
        if(count<=4):
            continue
        arr1=line.strip().split("\t")
        clade_name=arr1[0]
        tax_ids=arr1[1]
        abundance=arr1[2]
        clade_arr=clade_name.split('|')
        tax_arr=tax_ids.split('|')
        if(len(tax_arr)!=len(clade_arr)):
            print("the length of both is not equal")
        for i in range(len(clade_arr)):
            level=clade_arr[i]
            tax_id=tax_arr[i]
            if(taxlevel=="species"):
                if(level.startswith("s__")):
                    taxmap[int(tax_id)]=float(abundance)
            elif(taxlevel=="genus"):
                if(level.startswith("g__")):
                    if(tax_id==''):
                        continue
                    taxmap[int(tax_id)]=float(abundance)
    readfile.close()
    print(len(taxmap))
    print(taxmap)
    keys=set(taxmap.keys())
    print(keys)
    print("finished")
    return taxmap
```

```python
plt.rcParams['figure.figsize'] = (28, 22)
### 1：将数据集里分类正确的过滤出来。
def compute_average_error(TrueLevelcounter,Predict):
    ##计算average error
    n=len(TrueLevelcounter)
    #print(TrueLevelcounter)
    #print(Predict)
    result=0;
    for k,v in TrueLevelcounter.items():
        R_true=v;
        R_est=Predict[k]
        result=result+abs(R_true-R_est)/R_true
    return result/n

def AnalysisAbundance(df,level):
    
    ##统计答案的species的数量
    print("===========================TrueLevel=======================================")
    TrueLevel=list(df["true_"+level])
    TrueLevelcounter=dict(Counter(TrueLevel))
    if( -1 in TrueLevelcounter):
        del TrueLevelcounter[-1]
    import copy
    TrueLevel=copy.deepcopy(TrueLevelcounter) ##复本，留着计算average_error用
    keys_true=set(TrueLevelcounter.keys())
    print(TrueLevelcounter)
    
    ##统计基于cluster的level的数量
    print("===========================CluLevel=======================================")
    CluLevel=list(df["clu_"+level])
    CluLevelcounter=dict(Counter(CluLevel))
    if( -1 in CluLevelcounter):
        del CluLevelcounter[-1]
    if( 0 in CluLevelcounter):
        del CluLevelcounter[0]
    keys_clu=set(CluLevelcounter.keys())
    print(CluLevelcounter)

    
    ##统计kraken2的level的数量
    print("===========================KraLevel=======================================")
    LstKraLevel=list(df["kra_"+level])
    DictKraLevel=dict(Counter(LstKraLevel))
    KraLevel=dict()
    for key,val in DictKraLevel.items():
        if(val>10):
            KraLevel[key]=val
    if( -1 in KraLevel):
        del KraLevel[-1]
    if( 0 in KraLevel):
        del KraLevel[0]
    keys_kra=set(KraLevel.keys())
    print(KraLevel)
    
    print("==========================================================================")
    
    allset=keys_kra.union(keys_true).union(keys_clu)
    print("length pf all key: "+str(len(allset)))
    for key in allset:
        if key not in TrueLevelcounter:
            TrueLevelcounter[key]=0
        if key not in CluLevelcounter:
            CluLevelcounter[key]=0
        if key not in KraLevel:
            KraLevel[key]=0    
    allkeyset=list(allset)
    allkeyset.sort()
    print(allkeyset)
    
    values_true=[]
    values_clu=[]
    values_kra=[]
    for k in allkeyset:
        values_true.append(TrueLevelcounter[k])
        values_clu.append(CluLevelcounter[k])
        values_kra.append(KraLevel[k])
    ##计算
    Cluster_Avg_Err=compute_average_error(TrueLevel,CluLevelcounter)
    Kraken_Avg_Err=compute_average_error(TrueLevel,KraLevel)
    print("Cluster_Avg_Err: "+str(Cluster_Avg_Err))
    print("Kraken_Avg_Err: "+str(Kraken_Avg_Err))
```

```python
import pandas as pd
# -*- coding: utf-8 -*-
import numpy as np  
import matplotlib.mlab as mlab  
import matplotlib.pyplot as plt 
import math

plt.rcParams['figure.figsize'] = (34, 22)
def get_CAMI2_Distribution(inpath,labelpath,metapath):
    local_filter_df=pd.read_csv(inpath,sep='\t',header=None)
    local_filter_df.columns=['readname' ,'clusterid']
    print(local_filter_df.head())
    
    local_filter_readname=list(local_filter_df['readname'])
    local_filter_genome=list()
    
    
    labelfile=open(labelpath,'r')
    readLabelDict=dict()
    for line in labelfile:
        arr=line.strip().split("\t")
        readName=arr[0]
        labelId=arr[1]
        readLabelDict[readName]=labelId
    labelfile.close()
    print("readLabelDict")
    print(len(readLabelDict))
    
    metaFile=open(metapath,'r')
    labelNCBIID_Dict=dict()
    for line in metaFile:
        arr=line.strip().split("\t")
        label=arr[0]
        NCBI_ID=arr[2]
        labelNCBIID_Dict[label]=NCBI_ID
    metaFile.close()
    print("labelNCBIID_Dict")
    print(len(labelNCBIID_Dict))
    ##
    readNCBI_DICT=dict()
    for readName,labelId in readLabelDict.items():
        readNCBI_DICT[readName]=labelNCBIID_Dict[labelId]
    print("readNCBI_DICT")
    print(len(readNCBI_DICT))
    
    for readname in local_filter_readname:
        species_id=get_level(int(readNCBI_DICT[readname]),"species")
        #print(species_id)
        local_filter_genome.append(species_id)
    localfilter_readname_ncbi=dict(Counter(local_filter_genome))
    print(localfilter_readname_ncbi)
    print(len(localfilter_readname_ncbi))

    X=localfilter_readname_ncbi.keys()
    Y_pre=localfilter_readname_ncbi.values()
    Y=list()
    for value in Y_pre:
        Y.append(math.log10(value))
    fig = plt.figure()
    print(Y)
    plt.bar(X,Y,0.4,color="green")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("bar chart")

    plt.show()
```

## 分析reference sequence里taxid和 CAMI2的true species taxid的关系

```python
def analysis_refence(inpath):
    infile =open(inpath,'r')
    ref_set=set()
    for line in infile:
        fields=line.strip().split("\t")
        taxid=int(fields[1])
        ref_set.add(taxid)
    infile.close()
    return ref_set
```

```python
def analysis_refence_level(inpath,level):
    infile =open(inpath,'r')
    ref_set=set()
    for line in infile:
        fields=line.strip().split("\t")
        taxid=int(fields[1])
        taxid=get_level(int(taxid),level)
        ref_set.add(taxid)
    infile.close()
    return ref_set
```

```python
inpath="/home/ubuntu/kraken2/database2/seqid2taxid.map"
ref_set=analysis_refence(inpath)
```

```python
def get_rawdf(inpath,labelpath,metapath,clusterpath):
    nameset=set()
    count=0
    infile=open(inpath,'r')
    for line in infile:
        if(count%2==0):
            readname=line.strip().replace(">","")
            nameset.add(readname)
        count=count+1
    infile.close()
    
    ##读取答案文件
    labelfile=open(labelpath,'r')
    dict1=dict()
    for line in labelfile:
        arr=line.strip().split("\t")
        if arr[0] in nameset:
            dict1[arr[0]]=arr[1]
    labelfile.close()
    ser=pd.Series(dict1)
    df=pd.DataFrame(ser,index=None)
    df=df.reset_index(drop=False)
    df.columns=['readname','labelid']

    ##读取label到taxid的映射
    meta = pd.read_csv(metapath, sep="\t",usecols=[0,2])
    meta.columns=['labelid','taxid']
    df=pd.merge(df,meta,how='inner',on='labelid')
    
    ##读取local clustering到read的映射
    clu_df = pd.read_csv(clusterpath, sep="\t",header=None)
    clu_df.columns=['readname','clusterid']
    df=pd.merge(df,clu_df,how='inner',on='readname')
    df=df.sort_values(by=['readname'])
    print("=======================================打印前五行===================================================")
    print(df.head())
    print(df.shape)
    return df
```
